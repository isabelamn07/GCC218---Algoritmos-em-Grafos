<h3>Código</h3>
<p> O código apresentado a seguir utiliza lista de adjacência para representação do grafo, cada chave(key), representa um dos vértices presentes no grafo, e os items atrelados a cada chave, representam o vértice adjacente e se existir, o peso da aresta.

```python
### Grafos ###

# Grafo nao orientado, valorado
class Graph:
    def __init__(self):
        '''Construtor padrao
        Dicionario que representa a lista de adjacencia de um vertice
        formato: vertice_u: {vertice_v: peso}
        '''
        self.dict = { 
            1: {2: 4,  5: 9, 6: 8},
            2: {3: 6, 4: 6, 7: 12},
            3: {6: 23, 5: 11, 7: 55},
            4: {2: 4, 5: 99, 6: 46},
            5: {1: 7, 3: 6, 4: 32},
            6: {1: 25, 4: 14},
            7: {2: 44, 3: 78, 4: 19}
        }

    def eh_vizinho(self, u, v):
        '''
        Verifica adjacencia, se u for vizinho de v retorna True
        '''
        if u in self.dict:
            if v in self.dict[u]:
                return True
        else:
            return False
        return False

    def retorna_vizinhos(self, u):
        '''
        Retorna vertices adjacentes ao vertice de origem u dado
        '''
        if u in self.dict:
            return self.dict[u]
        else:
            print("Vertice nao encontrado!")

    def inserir_vertice(self, u, v, peso):
        '''
        Insere um vertice u caso o mesmo nao exista na lista de adjacencia
        '''
        if u not in self.dict:
            self.dict.update({u: {v: peso}})
        print(self.dict)
    
    def inserir_aresta(self, u, v, peso):
        '''
        Insere aresta dados dois vertices e seu peso
        '''
        if u and v in self.dict and v not in self.dict[u]:
            self.dict[u].update({v: peso})
            self.dict[v].update({u: peso})
        print(self.dict)
        

    def remover_vertice(self, u):
        '''
        Remove um dado vertice u
        '''
        if u in self.dict:
            self.dict.pop(u)
        else:
            print("Vertice nao encontrado")
        # Remove o vertice de toda a lista de adjacencia
        for i, j in self.dict.items():
            if u in j:
                j.pop(u)
        print(self.dict)

    def remover_aresta(self, u, v):
        '''
        Remove uma aresta, dado um vertice de origem e chegada
        '''
        if u in self.dict:
            if v in self.dict[u]:
                self.dict[u].pop(v)
                self.dict[v].pop(u)
            else:
                print('Vertice v nao encontrado')
        else:
            print('Vertice u nao encontrado')
        print(self.dict)

    def deletar_grafo(self):
        '''
        Deleta o grafo
        '''
        self.dict.clear()

    def main(self):
        print(self.eh_vizinho(1, 8))
        print(self.retorna_vizinhos(56))
        self.inserir_vertice(98, 6, 12)
        self.inserir_aresta(1, 7, 68)
        self.remover_vertice(6)
        self.remover_aresta(1, 12)
        

if __name__ == '__main__':
    graph = Graph()
    graph.main()
```
