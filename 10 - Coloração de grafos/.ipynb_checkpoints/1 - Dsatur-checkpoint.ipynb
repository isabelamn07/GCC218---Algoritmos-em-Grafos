# Algoritmo Dsatur


No algoritmo de DSATUR encontramos o vértice de maior grau e atribuimos a ele uma cor inicial, podemos representar as cores através de números, cada número representa uma cor. Depois procuramos o segundo vértice de maior grau, checamos suas adjacências e atribuimos a cor de menor valor já existente, caso o vértice seja ajacente a um vértice já colorido, e não existem cores disponíveis para atribuir, criamos uma nova cor e atribuimos o valor subjacente à última cor já utilizada.

### Pseudocódigo
1. Gere uma ordenação de **V(conjunto de todos os vértices)** em ordem decrescente a partir do grau dos vértices.
2. Selecione um vértice com grau maximal e atribuia a primeira cor.
4. Itere qualquer ordem para realizar a coloração dos outros vértices. Use a primeira possível na ordem de prioridades.
5. A execução termina quando **V** estiver vazio.

# Código 

```python
import networkx as nx
import matplotlib.pyplot as plt

#lista de adjacência
grafo = {'x1': {'x2':0, 'x3':0, 'x4':0, 'x5':0, 'x9':0},
         'x2': {'x1':0, 'x3':0, 'x4':0, 'x9':0, 'x10':0},
         'x3': {'x1':0, 'x2':0, 'x5':0},
         'x4': {'x1':0, 'x2':0, 'x5':0, 'x7':0},
         'x5': {'x1':0, 'x3':0, 'x4':0, 'x6':0, 'x7':0, 'x8':0},
         'x6': {'x5':0, 'x8':0, 'x9':0},
         'x7': {'x4':0, 'x5':0, 'x9':0, 'x10':0},
         'x8': {'x6':0, 'x5':0, 'x9':0},
         'x9': {'x1':0, 'x2':0, 'x6':0, 'x7':0, 'x8':0, 'x10':0},
         'x10': {'x7':0, 'x9':0, 'x2':0}
          }

# Dicionário que gera um índice para cada vértice
vertices = {}
# Lista que recebe os vértices do grafo de acordo com o grau em ordem decrescente
grau = []
count = 0 # contador que define o índice dos vértices no grafo
for i in grafo:    
    vertices.update({i: count})
    count += 1
        
for i in grafo:
    grau.append((len(grafo[i]), i))

V = list(grafo.keys()) # V recebe o conjunto de todos dos vértices do grafo

saturacao = {} # O dicionário saturacao recebe os vértices e suas respectivas cores

def dsatur():
    '''
    Função DSATUR, colore os vértices, a partir do vértice de maior grau.
    '''
    while V: # A coloração de vértices acontece enquanto V não estiver vazio
        maior_grau = max(grau) # maior_grau recebe o vértice de maior grau que ainda não foi colorido
        grau.remove(maior_grau) # o vértice precisa ser removido na lista de vértices e graus para não ser escolhido novamente
        if not saturacao:
            # Se o dicionário saturacao for vazio o primeiro vértice escolhido recebe a primeira core representada pelo número 1
            saturacao.update({1: [maior_grau[1]]})   
            V.remove(maior_grau[1]) # removemos os vértice já colorido de V
        else:
            satur(maior_grau[1]) # A funcao satur define a cor que o vertice ira receber
            V.remove(maior_grau[1])
    
def satur(maior):
    '''
    Função que analise os vértices adjacentes ao vertice a ser colorido
    
    Parametros:
        maior - vértice a ser colorido
    '''
    igual = [] # recebe as cores dos vertices adjacentes ao vertice a ser colorido
    flag = False
    for i in saturacao:
        if maior not in saturacao[i]:
            for j in range(len(saturacao[i])):
                if saturacao[i][j] in grafo[maior]:                     
                    igual.append((i))

    for j in saturacao:
        if j not in igual:
            saturacao[j].append(maior)
            return
        else:
            flag = True
    if flag:
        saturacao.update({len(saturacao)+1: [maior]})                 
    
dsatur()
print("Grafo colorido no formato {cor: [vertices]}: ", saturacao)

# Desenhando o grafo com networkx e matplotlib
G = nx.Graph(grafo)

plt.figure()
nx.draw(G, with_labels=True, font_weight='bold', node_size=700, font_size=18, width=5.5, font_color='w')
plt.show()
```
