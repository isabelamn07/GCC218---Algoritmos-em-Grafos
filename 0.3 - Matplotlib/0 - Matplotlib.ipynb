### Criação de uma figura

### figure()
Figure é o container em nível mais alto que contém os elementos necessários para plotar um gráfico. figure() é chamada por padrão, porém é possível chamá-la explicitamente para definir propriedades da figura a ser gerada.

Função: 
**matplotlib.figure.Figure(figsize=None, dpi=None, facecolor=None, edgecolor=None, linewidth=0.0, frameon=None, subplotpars=None, tight_layout=None, constrained_layout=None)**

Parâmetros:
* figsize - dimensões da figura.
* dpi - pontos por polegada.
* facecolor - cor de rosto do fragmento da imagem .
* edgecolor - cor das bordas do fragmeto da imagem.
* linewidth - a largura do frame da figura
* frameon - recebe valores booleanos. Se falso, remove o desenho de fundo do fragmento
* subplotpars - parâmetros do subplot
* tight_layout - se falso, utiliza subplotpars. Se for verdadeiro, ajusta os parâmetros do subplot usando tight_layout com espaçamento padrão.
* constrained_layout - se verdadeiro, é usado para para ajustar o posicionamento dos elementos da plotagem.



### subplots()

Cria a figura e um conjunto de subplots

Parâmetros:
* nrows - número de linhas
* ncols - número de colunas
    
OBS.: possui outros parâmetros para explicação de cada um deles - [função subplots](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots.html).

### Primeiro exemplo

````python
import matplotlib.pyplot as plt

fig = plt.figure(3, [7, 4.8], 100.0, 'b', 'r', constrained_layout=True)

fig, ax_list = plt.subplots(2,3)
````

### plot()

Função para plotar linhas ou marcadores como **(x, y)**.

Com a função **plot()**, podemos criar grafos a partir de listas. Essas listas representam os eixos **x** e **y** respectivamente.

Parâmetros:
* x, y - valores de x e y
* fmt - uma string que define que tipo de marcador ou linha representará cada ponto **(x,y)**, assim com as cores. [Lista completa de marcadores e cores](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html)

OBS.: existem vários outros parâmetros, você pode encontrá-los na documentação - [Documentação da função plot()](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html)    

````python
list1 = [1, 2, 3, 4, 5, 6, 7]
listy = [10, 21, 33, 89, 45, 9, 3]
listy2 = [56, 13, 8, 14, 56, 66, 44]
plt.plot(list1, listy)
plt.plot(list1, listy2)
````

## Funções
**xlabel()** - Define o título do eixo x.

**ylabel()** - Define o título do eixo y.

**title()** - Define o título do gráfico.

**legend()** - O comportamento e atributos de uma legenda inserida no gráfico.

**gride()** - Insere um grid no grafo.

**show()** - Mostra uma figura. No jupyter lab ou no jupyter notebook, a figura aparecerá independentemente da utilização da função **show()**. Porém, ao desenvolver em outro ambiente, é necessário utilizá-la para plotar os gráficos.


````python
plt.xlabel("Lista1")
plt.ylabel("Lista2 e 3")
plt.title("Teste")
list1 = [1, 2, 3, 4, 5, 6, 7]
listy = [29, 21, 33, 76, 45, 27, 55]
listy2 = [56, 13, 8, 14, 56, 66, 44]
plt.plot(list1, listy, label="avg")
plt.plot(list1, listy2, label="avg2")

plt.legend(loc="lower right", shadow="true")
plt.grid()
````

### axis()
Função que define os valores máximos e mínimos de **x** e **y**.

Parâmetros:
* xmin, xmax, ymin, ymax - limites dos valores dos eixos.
* option - se for um valor booleano, determina se linhas e labels serão mostrados ou não, se for uma string determina outras propriedades: [parâmetros da função axis()](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.axis.html).


````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 'ro' plota círculos vermelhos
plt.plot([3, 4, 5], [17, 5, 11], 'bo')  # 'bo' plota círculos azuis
plt.axis([0, 6, 0, 20])
# plt.axis([xmin, xmax, ymin, ymax])
plt.show()
````

````python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

````python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

````python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
````

````python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
````
