<h2>Travel Salesman Problem ou Problema do caixeiro viajante</h2>
<p>O problema do caxeiro viajante consiste em encontar o menor trajeto entre vários pontos, Passamos por cada um desses pontos(vértices) somente uma vez, e o ponto de partida também será o ponto de chegada. </p>

Suponha que um caxeiro viajante precise passar por cidades de Minas Gerais. Porém, ele precisa traçar um caminho que otimize não só o tempo que ele vai gastar, mas também suas despesas com a viagem.


Infelizmente ainda não foi descoberto um algoritmo em tempo polinomial que resolva o problema do caxeiro viajante e gere uma solução ótima. Temos que trabalhar com heusrísticas para encontrar soluções. Existem diversas formas de se encontrar um caminho para o problema do caxeiro viajante, veremos alguns desses algoritmos.
