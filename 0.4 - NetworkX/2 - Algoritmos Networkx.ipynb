# Algoritmos

Networkx tem vários algoritmos sobre teoria de grafos implementados. Encontre a lista desses algoritmos no link: [Lista de algoritmos networkx](https://networkx.github.io/documentation/stable/reference/algorithms/index.html)

Para cada tópico dessa biblioteca, são encontradas diversas funções. Porém, a lista de algoritmos implementados é muito extensa. Aqui veremos apenas alguns exemplos.

### Grafo bipartido
Entre as diversas funções no tópico sobre grafos bipartidos, temos a função **is_bipartite()**, que verifica se um grafo é bipartido ou não.

* Parâmetro
    - G(grafo)

```python
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt


B = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
B.add_nodes_from(vertices)
arestas = [('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'E'), ('C', 'F'), ('D', 'E'), ('D', 'F'), ('E', 'G'), ('F', 'G')]
B.add_edges_from(arestas)
print(bipartite.is_bipartite(B))

B.add_edge('D', 'G')
print(bipartite.is_bipartite(B))

# Desenhando o grafo com networkx
plt.figure()
nx.draw(B, with_labels=True, font_weight='bold', node_size=700, font_size=18, width=5.5, font_color='w')
plt.show()
```

### Pontes
A função bridges() detecta pontes em um grafo: 

* Parâmetro
    - G(grafo)

```python
import networkx as nx
import matplotlib.pyplot as plt

C = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O']
C.add_nodes_from(vertices)
arestas = [('A', 'B'), ('A', 'C'), ('A', 'E'), ('B', 'D'), ('B', 'F'), ('D', 'E'), ('E', 'K'), ('F', 'G'), ('F', 'J'),
           ('G', 'H'), ('G', 'I'), ('H', 'I'), ('J', 'K'), ('K', 'L'), ('L', 'N'), ('L', 'O'), ('M', 'N'), ('N', 'O')]
C.add_edges_from(arestas)
print("Pontes no grafo: ",list(nx.bridges(C)))

# Desenhando o grafo com networkx
plt.figure()
nx.draw(C, with_labels=True, font_weight='bold', node_size=700, font_size=18, width=5.5, font_color='w')
plt.show()
```

### Coloração
A função coloring.greedy_color() colore grafos utilizando algoritmos gulosos, tais como: DSATUR, maior saturação e sequência aleatória:

* Parâmetros
    - G(grafo);
    - stategy(string) - lista de estratégias possíveis: largest_first, random_sequential, smallest_last, independent_set, connected_sequential_bfs, connected_sequential_dfs, connected_sequential, saturation_largest_first, DSATUR.

```python
import networkx as nx
import matplotlib.pyplot as plt

D = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O']
D.add_nodes_from(vertices)
arestas = [('A', 'B'), ('A', 'C'), ('A', 'E'), ('B', 'D'), ('B', 'F'), ('D', 'E'), ('E', 'K'), ('F', 'G'), ('F', 'J'),
           ('G', 'H'), ('G', 'I'), ('H', 'I'), ('J', 'K'), ('K', 'L'), ('L', 'N'), ('L', 'O'), ('M', 'N'), ('N', 'O')]
D.add_edges_from(arestas)

a = nx.coloring.greedy_color(D, strategy='largest_first')
print(a)

b = nx.coloring.greedy_color(D, strategy='random_sequential')
print(b)

c = nx.coloring.greedy_color(D, strategy='smallest_last')
print(c)

d = nx.coloring.greedy_color(D, strategy='independent_set')
print(d)

e = nx.coloring.greedy_color(D, strategy='connected_sequential_bfs')
print(e)

f = nx.coloring.greedy_color(D, strategy='connected_sequential_dfs')
print(f)

g = nx.coloring.greedy_color(D, strategy='connected_sequential')
print(g)

h = nx.coloring.greedy_color(D, strategy='saturation_largest_first')
print(h)

i = nx.coloring.greedy_color(D, strategy='DSATUR')
print(i)
```

### Grafo Euleriano
Podemos verificar se um grafo é euleriano ou não com a função **is_eulerian()**.
* Parâmetro:
    - G(grafo).

```python
E = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
E.add_nodes_from(vertices)
arestas = [('A', 'B'), ('A', 'C'), ('A', 'D'), ('A', 'F'), ('B', 'E'), ('C', 'F'), ('D', 'G'), ('E', 'F'), ('F', 'G')]
E.add_edges_from(arestas)

print(nx.is_eulerian(E))
```

É possível também euleriar um grafo com a função eulerize().
* Parâmetro:
    - G(grafo).

```python
import matplotlib.pyplot as plt
import networkx as nx

F = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
F.add_nodes_from(vertices)
arestas = [('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'E'), ('C', 'F'), ('D', 'E'), ('D', 'F'), ('E', 'G'), ('F', 'G')]
F.add_edges_from(arestas)
print(F.edges, F.nodes)
print(nx.is_eulerian(F))

G = nx.eulerize(F)
print(nx.is_eulerian(G))
print(G.edges, G.nodes)

# networkx 
plt.figure(figsize = [6.4, 6.0], constrained_layout=True)
nx.draw(G, with_labels=True, font_weight='bold', node_size=700, font_size=18, width=5.5, font_color='w')
plt.show()
```

### Caminho mínimo
Podemos encontrar o caminho mínimo em um grafo utilizando uma das funções dos algorimos apresentado no networkx. Aqui, é utilizada a função **shortest_path()** 
* Parâmetros
    - G(grafo);
    - source(vertice) - vértice inicial para encontar o caminho mínimo;
    - target(vertice) - vértice de final para encontar o caminoh mínimo.

```python
import networkx as nx
import matplotlib.pyplot as plt

H = nx.Graph()
vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O']
H.add_nodes_from(vertices)
arestas = [('A', 'B'), ('A', 'C'), ('A', 'E'), ('B', 'D'), ('B', 'F'), ('D', 'E'), ('E', 'K'), ('F', 'G'), ('F', 'J'),
           ('G', 'H'), ('G', 'I'), ('H', 'I'), ('J', 'K'), ('K', 'L'), ('L', 'N'), ('L', 'O'), ('M', 'N'), ('N', 'O')]
H.add_edges_from(arestas)
print("Caminho mínimo definindo source e target: ", nx.shortest_path(H, 'A', 'J')) # com definição de vértice inicial e vértice final
print("Caminho mínimo: ", nx.shortest_path(H)) # sem definição de vértice inicial ou final
print("Caminho mínimo definindo source: ", nx.shortest_path(H, source='D')) # definindo apenas o vértice inicial
print("Caminho mínimo definindo target: ", nx.shortest_path(H, target='J')) # definindo apenas o vértice final

plt.figure(figsize = [6.4, 6.0], constrained_layout=True)

pos = nx.spring_layout(H)  

# vertices
nx.draw_networkx_nodes(H, pos, node_size=900)

# arestas
nx.draw_networkx_edges(H, pos, edge_size=500, width=6)

# labels
nx.draw_networkx_labels(H, pos, font_size=20, font_family='sans-serif', font_weight='bold', font_color='w')

plt.axis('off')
plt.show()
```
