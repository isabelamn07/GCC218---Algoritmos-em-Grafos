# Desenhando grafos utilizando Matplotlib e NetworkX

Vamos desenhar os grafos gerados utilizando **Networkx** e a biblioteca **Matplotlib**.

### draw()

Para desenhar o grafo, podemos utilizar o **draw()**. Existem vários parâmetros para definir propriedades do grafo gerado, são eles:

* Parâmetros:
    - G(graph, obrigatório) - um grafo networkx.
    - pos(dicionário, opcional) - Um dicionário com vértices e posições como valores. Se não especificado, um posicionamento layout spring será computado.    
    - ax (objeto Matplotlib Axes, opcional) - Desenha o grafo em um eixo especificado do Matplotlib.
    - hold(boleano, opcional) - Configura o modo de espera do Matplotlib.
    - arrows(boleano, opcional(default=True))- Para grafos direcionados, se verdadeiro desenha setas.
    - arrowstyle(string, opcional(default='-|>')) - define o estilo das setas.
    - arrowsize(int, opcional(default=10)) - Para grafos direcionados, determina o tamanho da ponta da seta.
    - with_labels(int, opcional(default-True)) - desenha o nome dos vértices.
    - ax(objeto Matplotlib Axes, opcional) - desenha o grafo em eixos Matplotlib definidos.
    - nodelist(lista, opcional(default G.nodes())) - Desenha apenas vértices definidos.
    - edgelist(lista, opcional(default=G.edges())) - Desenha apenas arestas definidas.
    - node_size(scalar ou array, opcional(default-300)) - Tamanho dos vértices. Caso seja uma array, precisa ser do mesmo tamanho que nodelist.
    - node_shape(string, opcional(default-'o')) - A forma do vértice. 
    - alpha(float, opcional(default=None)) - A transparência dos vértices e arestas.
    - cmap(Matplotlib colormap, opcional(default=None)) - Mapeamento da intensidade das cores dos vértices
    - vmin, vmax(float, opcional(default=None))- Mínimo e máximo para a escala do mapeamento de cores dos vértices.
    - linewidths([None|scalar|sequence]) - Largura da linha da borda do símbolo (padrão = 1.0).
    - width(float, opcional(default=1.0)) - Largura da linha das arestas.
    - edge_color(cor ou array de cores, opcional(default='k')) - Cor da aresta.
    - edge_cmap(Matplotlib colormap, opcional(default-None)) - Mapa de cores para mapear intensidades de arestas.
    - edge_vmin, edge_vamx(float, opcional(defaullt=None)) - Mínimo e máximo para escala do mapeamento das cores das arestas.
    - style(string, opcional(default='solid')) - Estilo das arestas (solid|dashed(tracejada)|dotted(de pontos)|dashdot) 
    - labels(dicionário, opcional(default-None)) - Nome dos vértices em um dicionário com chaves que representam os textos dos vértices.
    - font_size(int, opcional(default=12)) = Tamanho da fonte.
    - font_color(string, opcional (default='k'black)) - Cor dos textos.
    - font_weight(string, opcional(default='normal')) - Largura dos texto.
    - font_family(string, opcional(default='sans-serif')) - Tipo de texto.
    - label(string, opcional) - Texto para legenda do grafo.

### Exemplo

```python
import matplotlib.pyplot as plt
import networkx as nx
G=nx.dodecahedral_graph()
nx.path_graph(20)
nx.draw(G, pos=nx.spring_layout(G), with_labels=True, node_size=800, node_color='#c235e6', alpha=0.89, node_shape='8', 
        cmap=8.0, vmin=3.0, vmax=9.0, linewidths=5.0, width=6.5, edge_color='#385de8',
        style='dashed', label='Grafo teste', font_size=20, font_color='w', font_weight='bold', font_family='arial')
plt.draw()

# Fonte: https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.drawing.nx_pylab.draw_networkx_labels.html
```

### draw_networkx_edges(), draw_networkx_nodes(), draw_networkx_labels(), draw_networkx_edge_labels()

Utiliza os mesmos parâmetros que são utilizado no **draw()**, porém somentes aqueles relacionados a sua respectiva função: 

* Parâmetros **draw_network_edges()**:
    - G(grafo);
    - pos (dicionário);
    - edgelist (coleção de tuplas de arestas);
    - width (float,ou array de floats);        
    - edge_color (string da cor, ou array de floats); 
    - style (string);
    - alpha (float); 
    - cmap (aresta);
    - edge_vmin,edge_vmax (floats); 
    - ax (objeto Matplotlib Axes); 
    - arrows (bool, (default=True));
    - label ([None| string]). 
* Parâmetros **draw_network_nodes()**:
    - G(grafo);
    - pos (dicionário);
    - ax (objeto Matplotlib Axes); 
    - nodelist (lista);
    - node_size (scalar ou array);
    - node_color(string da cor, ou array de floats);
    - node_shape (string);  
    - alpha (float); 
    - cmap (Matplotlib colormap);
    - vmin,vmax (floats); 
    - linewidths (bool, (default=True));
    - label ([None| string]) .
* Parâmetros **draw_networkx_labels()**:
    - G(grafo);
    - pos (dicionário);     
    - labels (dicionário (default=None));
    - font_size (int);
    - font_color (string); 
    - font_family (string);
    - font_weight (string);
    - alpha (float);
    - ax (objeto Matplotlib Axes). 
* Parâmetros **draw_networkx_edges_labels()**:
    - G(grafo);
    - pos (dicionário);
    - ax (objeto Matplotlib Axes);
    - alpha (float);
    - edge_labels(dicionário);
    - label_pos (float) - posição do label da aresta.(0=cabeça, 0.5=centro, 1=fim);
    - font_size (int);
    - font_color (string);
    - font_family (string);
    - font_weight (string);
    - bbox (Matplotlib bbox) - especifica texto, formas e cores;
    - clip_on(bool) - ativa corte nos limites dos eixos (default=True).

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.petersen_graph()
plt.figure(constrained_layout=True)
plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold', node_size=700, edge_size=500, node_color='#b5042d', font_size=18, width=5.5, font_color='w')
plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold', node_size=700, font_size=18, width=5.5, font_color='w')
plt.show()

# Fonte: https://networkx.github.io/documentation/stable/tutorial.html
```

```python
# Author: Aric Hagberg (hagberg@lanl.gov)
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

G.add_edge('a', 'b', weight=0.6)
G.add_edge('a', 'c', weight=0.2)
G.add_edge('c', 'd', weight=0.1)
G.add_edge('c', 'e', weight=0.7)
G.add_edge('c', 'f', weight=0.9)
G.add_edge('a', 'd', weight=0.3)

elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.5]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 0.5]

pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge,
                       width=6)
nx.draw_networkx_edges(G, pos, edgelist=esmall,
                       width=6, alpha=0.5, edge_color='b', style='dashed')

# labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

plt.axis('off')
plt.show()

# Fonte: https://networkx.github.io/documentation/stable/auto_examples/drawing/plot_weighted_graph.html
```
